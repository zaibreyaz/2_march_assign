```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot."""

"""Ans: A scatter plot is a graph that displays two sets of numerical data as a collection of points, with one variable on the x-axis and the other on the y-axis. 
        It's used to show the relationship between the two variables and to identify any patterns or trends in the data."""

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""

"""Ans: The subplot() function is used in Matplotlib to create multiple plots in the same figure. 
        It allows you to specify the number of rows and columns of subplots you want to create, as well as the position of each subplot within the figure."""

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

plt.show()
```

```python
"""Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot."""

"""Ans: A bar plot is a type of chart or graph that displays categorical data using rectangular bars. 
        The length of each bar represents the value of the category being displayed. Bar plots are used to compare and visualize the numerical values of different categories.
        Bar plots are commonly used in data analysis and visualization to:
            1. Show the frequency or count of each category in a dataset.
            2. Compare values between different categories in a dataset.
            3. Display the change or trend of values over time or across categories.
            4. Present data in a clear and easy-to-understand manner.
"""

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Bar Plot')
plt.show()
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company, profit)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Horizontal Bar Plot')
plt.show()
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot."""

"""Ans: A box plot is a type of graph that shows the distribution of a set of data. It's useful for identifying the range of values, outliers, and any patterns in the data. 
        It's commonly used in data analysis to compare different sets of data and to find important characteristics."""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
plt.ylabel('Y axis')
plt.title('Box Plot')
plt.show()
```
